```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.rename(columns=({'gender':'Gender','race/ethnicity':'Race_Eth'
                     ,'parental level of education':'PLEC'
                     ,'lunch':'Lunch','test preparation course':'TPC'
                      ,'math score':'MS','reading score':'RS'
                     ,'writing score':'WS'}),inplace=True)
```

```python
df.head()
```

```python
df.columns
```

Q1) Depict Race_Eth on a countplot

```python
d1=df['Race_Eth'].value_counts()
d1
```

```python
#by using bar chart
plt.bar(d1.index,d1.values)
plt.show()
```

```python
#by using seaborn
sns.countplot(x=df['Race_Eth'])
plt.grid()
plt.show()
```

```python
sns.countplot(x=df['Race_Eth'],palette=['orange','red','blue','green','purple'])
plt.grid()
plt.show()
```

```python
sns.countplot(x=df['Race_Eth'],color='orange')
plt.grid()
plt.show()
```


Q2) Depict Gender on a counplot
